Since `A` is nonsingular (fullrank), we can apply the QR factorization.
In QR factorization the given matrix `A` can be represented as the product of an orthogonal matrix `Q` and an upper triangular matrix `R`.
Since the Scipy method for the QR factorization is based on the modified Gram-Schmidt method, the matrix `Q` will have columns that form an othongonal basis for the column space of `A` that is already normalized. #### Comment:
Latent Semantic Index is a retrieval method that uses SVD to identify patterns in the ralationships between the terms and the concepts contained in the document collection.LSI is based on the principle that words that are used in the same context tend to have similar meanings.

This algorithm consists of three main steps:
- Calculate the SVD of the matrix terms by documents
- Select a suitable k to obtain a valid truncated SVD
- Calculate the cosine similarity between the query and the new approximate matrix.

```python
# Full_matrices = false -> allows us to not consider the zero singular values
U, S, Vh = spl.svd(An, full_matrices=False)
```

```python
print('Shape of U: ', U.shape)
print('Shape of S: ', S.shape)
print('Shape of V^T: ', Vh.shape)
```

```python
#Scree plot method
plt.semilogy(S,'o')
plt.title('Scree plot of Singular Values')
plt.xlabel('Index')
plt.ylabel('Singular Value')
plt.show()
```

We can visualize that the first 3 singular values alone give us much of the information, but it is difficult to establish, by the Scree plot, where the elbow is placed (that is, how many singular values we should consider for approximation). Maybe consider a about 10 singular values might be a good compromise.
Anyway, let us use three other methods to confirm or improve our intuition: the cumulative variance method, the Keiser Rule and the relative error.

The cumulative variance method was chosen because the treshold to set is known. Keiser Rule was chosen because it does not need a treshold and the relative error method was chosen because it returns the lower value for k.

To define the optimal number of components $k$ using cumulative variance method, we will use the formula

$$E_k = \frac{\sum_{i=1}^{k} \sigma_i^2}{\sum_{j=1}^{n} \sigma_j^2}$$

It is necessary to choose a cut-off $E^*$, usually between 70 and 80% and we retain the first k directions such that $E_k>E^*$ and $k$ is as small as possible.

```python
# Cumulative variance method
cumulative_sum_k = np.cumsum(S**2)
total_sum = np.sum(S**2)

threshold = 0.7

for k in range(1, len(S) + 1):
    Ek = cumulative_sum_k[k - 1] / total_sum
    if Ek >= threshold:
        break

print('Desired cumulative variance: ', threshold)
print('Smallest k: ', k)
print('Value of Ek: ', Ek)
```

To define the optimal number of components $k$ using the Keiser Rule, we will use the formula

$$ \bar \sigma = \frac{1}{r}\sum_{i=1}^r \sigma _i^2 $$

to compute the mean variance.

Then, we retain the number of singular values whose variance exceed $\bar \sigma $, i.e. $ \sigma _k^2 > \bar \sigma$.

```python
# Kaiser rule method
# Compute mean variance
sigma_mean = np.mean(S**2)

k_kaiser = np.sum(S**2 > sigma_mean)

print('Mean variance', sigma_mean)
print('Smallest k with Kaiser rule: ', k_kaiser)
```

To define the optimal number of components $k$ using the Relative Error, according to Eckart Young theorem,  we will use the formula

$$ \frac{\lVert A-A_k\rVert _2}{\lVert A \rVert _2} = \frac{\sigma_{k+1}}{\sigma_1} \leq tol $$

Then we will keep the k that satisfies the inequality.

Note: The closer the ratio is to zero, the better the approximation is. In fact it means that the approximate matrix coincides with the original matrix. As always, also in this case our goal is to take the smallest value of k for which the ratio remains small. I will choose a tolerance of $5e-1$ as it is in the middle.

```python
# Relative error method
tol = 5e-1

# Ratio between sigma_{k+1} and sigma_1 for each k
ratios = S[1:] / S[0]

plt.plot(range(1, len(ratios)+1), ratios, marker='o', linestyle='-', color='b')
plt.axhline(y=1-tol, color='r', linestyle='--', label='Threshold')

plt.xlabel('k')
plt.ylabel('Ratio')
plt.title('Relative Error Method')
plt.legend()

plt.show()

#for k in range(1, len(ratios) + 1):
    #ratio = ratios[k - 1]
    #print('k =', k, ', ratio:', ratio)

# Find the value of k where the ratio is less than or equal to tol
k = np.argmax(ratios <= 1-tol) + 1

print('\nThe value of k where ratio <= ', 1-tol, 'is:', k)
```

As we can see, all methods recommend suggest a k of just under 10. Therefore we will consider a k=8 given by the average of the suggested values.

So, we are dropping a generous number of SVD components because variation in the ambiguity of many words produces significant noise in A. Coupling this with the fact that numerical accuracy in not an important issue menas that we are happy to replace the SVD of A by low-rank truncation $A_k$, where k is significantly less than $r$. (Meyer p.420)

Then we consider $k=8$ and build the approximated matrix Ak

Now, we have that $A = USVh$ is the SVD of the term-document matrix and, according to the Low Rank Approximation theorem, we can approximate it by a matrix of rank $k$ obtaining that $$A  U_k(S_kV_k^T) := U_k D_k$$

```python
# According to error relative method
k = 8

# Truncate SVD to retain only the top k components
Uk = np.copy(U[:, :k])
print('Shape of Uk: ', Uk.shape)

Sk = np.copy(np.diag(S[:k]))
print('Shape of Sk: ', Sk.shape)

Vhk = np.copy(Vh[:k, :])
print('Shape of V^Tk: ', Vhk.shape)
```

The columns of $U_k$ are an orthogonal basis that we use to approximate the documents.

From $A  U_k D_k$ we have $a_j  U_k d_j$ , which means that
column j of $D_k$ holds the coordinates of document j in terms of the orthogonal
basis.


```python
#print(np.linalg.norm((A-Ak),2))
#print(np.linalg.norm((A),2))
#print(np.linalg.norm((A-Ak),2)/np.linalg.norm((A),2))
```

```python
Dk = np.dot(Sk, Vhk)
Ak = np.dot(Uk, Dk)
print('Shape of Ak: ', Ak.shape)
```

```python
rank_Ak = np.linalg.matrix_rank(Ak)

print("Rank of Ak with", k, "singular values:", rank_Ak)
```

With this rank-k approximation the term-document matrix is represented
by $A_k = U_k D_k$ and in query matching we compute $$ q^T A_k = q^T U_k D_k = (U^T_k q)^T D_k $$

Thus, we compute the coordinates of the query in terms of the new document basis and compute the cosines from

$$ cos\theta _j = \frac{q_k^T d_j}{\lVert q_k\rVert_2 \lVert d_j \rVert_2} $$

with $q_k = U_k^T q$.

This means that the query-matching is performed in a k-dimensional space.

Furthermore, the vectors in $U_k$ and in $D_k$ only need to be computed once and they can be determined without computing the entire SVD, so the calculus of the cosine requires very little computation to process each new query. This observation will be useful for one of the following tasks.

Let's quantify, now, the difference between the cosine obtained with the complete matrix and the LSI. To do this we apply three measures: the standard deviation, the means of the squared error and the cosine. The ideal value for the first two measurements is 0, while for the calculation of cosine the ideal value is 1.

```python
difference = cosin_Theta_An - lsi_qk

std_dev = np.std(difference)

mse = np.mean((difference)**2)

cosine_theta = np.dot(lsi_qk.T, cosin_Theta_An) / (np.linalg.norm(lsi_qk,2) * np.linalg.norm(cosin_Theta_An,2))
```

The standard deviation is a measure of data dispersion, that is, how much the values of the two vectors deviate from their mean. Since the difference between the two vectors is zero if the vectors coincide, then the closer the standard deviation approaches zero the more similar the vectors are.

In this case I think the standard deviation is reasonably small to motivate a truncation with k=8.

The mean squared error is a measure of the average squared difference between the values of the two vectors. A higher value indicates greater discrepancy between the two vectors. For the same reasons above, if the vectors are very similar, the average of the difference vector will tend to zero and therefore the average of the square errors.

The cosine of the angle between the two vectors can can take values from -1 to 1. When the cosine is close to 1, the vectors are aligned, when close to 0, the vectors are orthogonal, when close to -1, the vectors are opposite. Therefore, a value of $0.658$ suggests that the vectors have a certain degree of alignment.

### Compute:
- Using the singular value decomposition, compute the closest point  to the query in the range of $A$ and in the range of $A_k$.

### Comment

The Closest Point theorem says that given a subspace $M$ of an inner product space $V$ and a vector $b$, the unique vector in $M$ that is closest to $b$ is $p = P_Mb$, the orthogonal projection of $b$ onto $M$.

Then, we need of an orthogonal projector on the column space of A ($P_{R(A)}$) and another one on the column space of Ak ($P_{R(Ak)}$). Such projector, generally, have the form: $P_{R(A)} = B(B^TB)^{-1} B^T$ and $P_{R(A^T)} = I - P_{R(A)}$, where $B$ is a matrix whose columns are bases for $R(A)$.

But we are lucky and we don't need to compute the inverse. In fact a way to describe the orthogonal projectors onto each of the four fundamental subspace is to make use of the pseudoinverse. Computing the products $AA^{-1}$ and $A^{-1} A$ replacing A with his SVD reveals that $AA^{-1} = U_1 U_1^T =: P_{R(A)}$ and $A^{-1} A = V_1 V_1^T =: P_{R(A^T)}$, where $U_1$ and $V_1$ constitute orthonormal bases for $R(A)$ and $R(A^T)$ respectively.

Furthermore, we already know how many vectors to take from $A$ to get a base for $R(A)$ and a base for $R(Ak)$, since we know both the rank of $A$ and $Ak$.
The rank of $A$ is 23 and the rank of $Ak$ is 8, so we can simply consider $U$ and $Uk$ respectively.

```python
#Compute the projectors
projector_RA = np.dot(U, U.T)
projector_RAk = np.dot(Uk, Uk.T)

#Compute the closest points
closest_point_A = np.dot(projector_RA, query1.T)
closest_point_Ak = np.dot(projector_RAk, query1.T)
```

```python
#Compute the orthogonal distances
orth_dist_A = np.linalg.norm((query1 - closest_point_A),2)

#Compute the orthogonal distances
orth_dist_Ak = np.linalg.norm((query1 - closest_point_Ak),2)

print(orth_dist_A)
print(orth_dist_Ak)
```

The similarity in orthogonal distances suggests that the dimensionality reduction performed by $Ak$ effectively preserves the main structure of the relationship between the columns of A, resulting in similar projections of the query in both cases.

- Construct the column-covariance matrix, numerically determine if it is positive definite. Do the eigenvalues of such a matrix satisfy the theoretical relation with the singular value of the matrix $A$?

The covariance indicates the level to which two variables vary together. The covariance matrix of a matrix $A_{mxn}$ is defined as $ C=\frac{X^T X}{m} $, where $X$ is the centered version of $A$, i.e. for each columns $x$ in $X$: $x_i=a_i-\mu (a_i)$ with $\mu$ the mean. Replacing $A$ with its SVD into the definition of the covariance matrix, we get: $$C=\frac{X^T X}{m}=\frac{(USV^T)^T USV^T}{m}=\frac{V\Sigma^2 V^T}{m}$$

We will see how to use this equivalence.

Note: Since the number of rows is small, the denominator will be $m-1$, then the covariance matrix will be unbiased.

```python
Cov = np.dot(X.T,X)/(A.shape[0]-1)
```

In general, one way to see if a matrix is positive definite is to check if all its eigenvalues are positive. Since the singular values in the SVD are the square roots of the eigenvalues of the positive matrix $A^TA$, we can check if a matrix is positive definite by examining its singular values. Hence, if all squared singular values are positive, then the matrix is positive definite.

At this point, however, it is essential to note that by definition a covariance matrix is symmetric and a fundamental property of symmetric matrices is that the eigenvalues are equal to the singular values.
